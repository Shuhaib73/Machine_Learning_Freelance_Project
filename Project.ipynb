## 2. Use machine learning to create clusters of similar projects

```python
# Importing necessary libraries for clustering, data preprocessing, and dimensionality reduction

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler              
from sklearn.decomposition import PCA                         # PCA (Principal Component Analysis): Used for dimensionality reduction
```

```python
# Splitting the dataset into training and testing sets

# Importing the train_test_split function from scikit-learn

from sklearn.model_selection import train_test_split

# Splitting the data into training and testing sets with an 80% training size and a random seed of 4

xtrain, xtest = train_test_split(df_scaled, train_size=0.8, random_state=4)
```

```python
# Displaying the shapes of the training and testing sets

print(f"Training Set Shape: {xtrain.shape}")
print(f"Testing Set Shape: {xtest.shape}")
```

```python
# Performing Principal Component Analysis (PCA) for dimensionality reduction

# Importing PCA from scikit-learn

from sklearn.decomposition import PCA
pca = PCA(n_components=2)                        # Creating a PCA instance with 2 principal components
xtrain = pca.fit_transform(xtrain)               # Applying PCA transformation to the training set & testing set
xtest = pca.transform(xtest)
```

```python
# Calculating the total explained variance ratio after PCA

total_explained_variance = sum(pca.explained_variance_ratio_)
print(f"Total Explained Variance: {total_explained_variance:.2f}")
```

```python
# Determining the optimal number of clusters using the Elbow Method

# Initialize an empty list to store within-cluster sum of squares (WCSS) values

wcss_list = []

for i in range(1,11) :
    k_model = KMeans(n_clusters=i)
    k_model.fit(xtest)
    wcss = k_model.inertia_
    wcss_list.append(wcss)
```

```python
# Plotting the Elbow Method for Optimal K

plt.figure(figsize=(10,5))
plt.plot(range(1,11), wcss_list, marker='o', linestyle='--', markerfacecolor='#98FB98')

# Highlight the optimal K (e.g., using a red dot)

optimal_k = 4  # Replace with your optimal K value
plt.scatter(optimal_k, wcss_list[optimal_k - 1], c='red', marker='o', s=100, label=f'Optimal K = {optimal_k}')

plt.title('Elbow Method for Optimal K', fontweight='bold')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Within-cluster Sum of Squares)')
plt.plot()
```

```python
# Creating a K-Means clustering model with 4 clusters

model = KMeans(n_clusters=4)

# Fitting the model to the training data

model.fit(xtrain)

train_pred = model.predict(xtrain)                # Predicting cluster labels for the training data
pred = model.predict(xtest)                       # Predicting cluster labels for the testing data
```

```python
# Visualizing the Clustering Results for the Testing Dataset

plt.scatter(xtest[pred==0,0], xtest[pred==0,1], s=20, c='#D4CC47', label='Cluster-1')
plt.scatter(xtest[pred==1,0], xtest[pred==1,1], s=20, c='#912CEE', label='Cluster-2')
plt.scatter(xtest[pred==2,0], xtest[pred==2,1], s=20, c='#20B2AA', label='Cluster-3')
plt.scatter(xtest[pred==3,0], xtest[pred==3,1], s=20, c='#FFB6C1', label='Cluster-4')

# Plotting cluster centers

plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], s=50, c='#0A0A0A', label='Cluster Centers')
plt.title('Clusters of Similar Projects', color='#1E1E1E', fontsize=14, fontweight='bold')

# Removing grid lines

plt.grid(False)

plt.legend()
plt.show()
```

```python
# Visualizing the Clustering Results for the Training Dataset

plt.scatter(xtrain[train_pred==0,0], xtrain[train_pred==0,1], s=10, c='#D4CC47', label='Cluster-1')
plt.scatter(xtrain[train_pred==1,0], xtrain[train_pred==1,1], s=10, c='#912CEE', label='Cluster-2')
plt.scatter(xtrain[train_pred==2,0], xtrain[train_pred==2,1], s=10, c='#20B2AA', label='Cluster-3')
plt.scatter(xtrain[train_pred==3,0], xtrain[train_pred==3,1], s=10, c='#FFB6C1', label='Cluster-4')

# Plotting cluster centers

plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], s=50, c='#0A0A0A', label='Cluster Centers')
plt.title('Clusters of Similar Projects', color='#1E1E1E', fontsize=14, fontweight='bold')

# Removing grid lines
plt.grid(False)

plt.legend()
plt.show()
```

### Overall Insights:

#### The clustering analysis has allowed us to group projects into four distinct clusters based on their feature similarities. These clusters can provide valuable insights into the nature of the projects, potentially guiding decision-making processes or further analysis.

```python
# Splitting the data into training and testing sets (80% training, 30% testing)

from sklearn.model_selection import train_test_split    
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size=0.9, random_state=4)
```

```python
# Applying all the model together

from sklearn import metrics

# # 1. LinearRegression

linear = LinearRegression()
ln = linear.fit(xtrain, ytrain)
y_train_ln = ln.predict(xtrain)
y_pred_ln = ln.predict(xtest)
accuracy_train_ln = metrics.r2_score(ytrain, y_train_ln)
accuracy_test_ln = metrics.r2_score(ytest, y_pred_ln)
print(accuracy_train_ln)
print(accuracy_test_ln)


# 2. SVM

sv = SVR(kernel='poly')
sv_mod = sv.fit(xtrain, ytrain)
y_pred_sv = sv_mod.predict(xtest)
accuracy_sv = metrics.r2_score(ytest, y_pred_sv)
print(accuracy_sv)

# # 3. DecisionTreeRegressor

dt = DecisionTreeRegressor()
dt_mod = dt.fit(xtrain, ytrain)
y_train_dt = dt_mod.predict(xtrain)
y_pred_dt = dt_mod.predict(xtest)
accuracy_train_dt = metrics.r2_score(ytrain, y_train_dt)
accuracy_test_dt = metrics.r2_score(ytest, y_pred_dt)
print(accuracy_train_dt)
print(accuracy_test_dt)

# # 4. RandomForest

rm = RandomForestRegressor()
rm_mod = rm.fit(xtrain, ytrain)
y_train_rm = rm_mod.predict(xtrain)
y_pred_rm = rm_mod.predict(xtest)
accuracy_train_rm = metrics.r2_score(ytrain, y_train_rm)
accuracy_rm = metrics.r2_score(ytest, y_pred_rm)
print(accuracy_train_rm)
print(accuracy_rm)

# # 5. KNeighborsRegressor

km = KNeighborsRegressor()
km_mod = km.fit(xtrain, ytrain)
y_pred_km = km_mod.predict(xtest)
accuracy_km = metrics.r2_score(ytest, y_pred_km)
print(accuracy_km)

# # 5. XGBRegressor

xm = XGBRegressor()
xm_mod = xm.fit(xtrain, ytrain)
y_pred_xm = km_mod.predict(xtest)
accuracy_xm = metrics.r2_score(ytest, y_pred_xm)
print(accuracy_xm)

# # 6. GradientBoostingRegressor

gm = GradientBoostingRegressor()
gm_mod = gm.fit(xtrain, ytrain)
y_train_gm = gm_mod.predict(xtrain)
y_pred_gm = gm_mod.predict(xtest)
accuracy_train_gm = metrics.r2_score(ytrain, y_train_gm)
accuracy_test_gm = metrics.r2_score(ytest, y_pred_gm)
print(accuracy_train_gm)
print(accuracy_test_gm)


# # 7. Ridge Regression

from sklearn.linear_model import Ridge

rd = Ridge()
rd_mod = rd.fit(xtrain, ytrain)
y_train_rd = rd_mod.predict(xtrain)
y_pred_rd = rd_mod.predict(xtest)
accuracy_train_rd = metrics.r2_score(ytrain, y_train_rd)
accuracy_test_rd = metrics.r2_score(ytest, y_pred_rd)
print(accuracy_train_rd)
print(accuracy_test_rd)
```

```python
# Create a list of regression model names - the models which are giving the best results

list_models = ['LinearRegression','DecisionTreeRegressor','RandomForestRegressor','GradientBoostingRegressor','Ridge Regression']

# Calculate and round the corresponding r2 scores

r2_training_score = [round(accuracy_train_ln,3),round(accuracy_train_dt,2),round(accuracy_train_rm,2),round(accuracy_train_gm,2),round(accuracy_train_rd,2)]

r2_testing_score = [round(accuracy_test_ln,3),round(accuracy_test_dt,2),round(accuracy_rm,2),round(accuracy_test_gm,2),round(accuracy_test_rd,2)]

# Create a DataFrame to store the results

reg_df = pd.DataFrame({'Algorithms_used' : list_models, 'r2_training_score' : r2_training_score, 'r3_testing_score' : r2_testing_score})

# Sort the DataFrame by r2_score in descending order

reg_df = reg_df.sort_values(by=['r2_training_score','r3_testing_score'], ascending=False)
reg_df
```
